```python
# Q1 (a)

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\DELL\Downloads\customers-100.csv")
df = df.head()


country_count = df['Country'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(country_count.index, country_count.values, color='green')
plt.title("Number of Customers in top 10 Country")
plt.xlabel("Country")
plt.ylabel("Number of customers")
plt.xticks(rotation=10)
plt.tight_layout()
plt.show()

```

```python
#Q1 (b)
top_cities = df['City'].value_counts().nlargest(5)

plt.pie(top_cities.values, labels=top_cities.index, autopct='%1.1f%%')
plt.title("Percentage of customers in top 5 cities")
plt.show()
```

```python
#Q2 (a)
df["Total Customers"] = range(1, len(df) + 1)
plt.plot(df["Total Customers"])
plt.title("Cumulative Number of Customers")
plt.xlabel("Row ")
plt.ylabel("Total Customers")
plt.grid(True)
plt.show()


```

```python
# Q2 (b)
plt.barh(top_cities.index, top_cities.values, color='orange')
plt.title("Top 5 Cities (Horizontal Bar)")
plt.xlabel(" No. of Customers")
plt.show()
```
